Que 1:List Operations and Basic Statistics:
        Create a list of the first 20 natural numbers.
        Compute the sum of all even numbers in the list.
        Compute the product of all odd numbers in the list.
        Calculate the mean, median, and standard deviation of the list.

```python
import statistics
list=[i for i in range(1,21)]
even_sum=sum(i for i in list if i%2==0)
prod=1
for i in list:
    if i%2!=0:
        prod*=i
print("Sum of even numbers 1 to 20: ",even_sum)
print("Product of all numbers 1 to 20: ",prod)
mean=statistics.mean(list)
med=statistics.median(list)
std=statistics.stdev(list)
print("Mean: ",mean)
print("Median: ",med)
print("Standard deviation: ",std)
```

Que 3:Fibonacci Sequence:
        Write a Python function that takes a number n and returns a list of the first n Fibonacci numbers.
        Plot the Fibonacci sequence using a line graph.

```python
import matplotlib.pyplot as plt
n=int(input())
tn_1=1
tn_2=0
list=[0,1]
for i in range (n-2):
    tn=tn_1+tn_2
    list.append(tn)
    tn_2=tn_1
    tn_1=tn
print(list)
xaxis=[i for i in range(1,n+1)]
mylabels=['N=10']
plt.plot(xaxis,list,'o-.c',ms=5,mec='magenta',mfc='magenta',linewidth=2,label=mylabels)
plt.xlabel("x-axis")
plt.ylabel("Fibonacci number")
plt.legend()
plt.grid()
plt.show()
```

Que 5:User Input and Basic Statistics:
        Write a program to read a comma-separated list of numbers from the user (e.g., “10, 20, 30, 40”).
        Calculate the mean of the numbers.

        Plot a histogram of the numbers.

```python
import matplotlib.pyplot as plt

input_numbers = input("Enter a comma-separated list of numbers (e.g., 10, 20, 30, 40): ")

numbers = [int(i) for i in input_numbers.split(",")]

mean = sum(numbers) / len(numbers)
smallest = min(numbers)
largest = max(numbers)

print(f"Numbers: {numbers}")
print(f"Mean: {mean}")
print(f"Smallest number: {smallest}")
print(f"Largest number: {largest}")

plt.hist(numbers, bins=5, color='blue', edgecolor='black', alpha=0.7)
plt.title("Histogram of Numbers")
plt.xlabel("Number Range")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

Que 10:Character Frequency:
        Write a function that accepts a string and:
        Converts all characters to lowercase.
        Counts the frequency of each character.
        Returns a dictionary with characters as keys and their counts as values.
        Visualize the character frequencies using a bar chart.

```python
import matplotlib.pyplot as plt
def ch_frequency(string):
    list=[i.lower() for i in string]
    frequencies=[]
    chara=[]
    for space in list:
        if space==" ":
            list.remove(space)
    for char in list:
        if char not in chara:
            chara.append(char)
            frequencies.append(list.count(char))
    return chara,frequencies

input_string=input("Enter the string: ")        #Hello world, this is python.
characters,frequency=ch_frequency(input_string)
plt.bar(characters,frequency,color="blue",label=input_string)
plt.xlabel("Character",size=15)
plt.ylabel("Frequency",size=15)
plt.legend()
plt.show()
```

Que 15:Anagram Finder:
        Write a Python function to find all anagrams of a word in a given list.
        Input: word = "listen", words = ["enlist", "google", "inlets", "banana"]
        Output: ["enlist", "inlets"]

```python
def anagram_finder(word,words):
    word_list=sorted([letter for letter in word])
    words_list=[]
    anagrams=[]
    for i in words:
        words_list.append({i:sorted([letter for letter in i])})
    for i in words_list:
        for j,k in i.items():
            if k==word_list:
                anagrams.append(j)
    return anagrams

word="listen"
words=["enlist", "google", "inlets", "banana"]
print(anagram_finder(word,words))

```

Que 16:Set Operations:
        Given two lists, write Python code to find their intersection, union, and symmetric difference.
        Visualize the results using Venn diagrams.

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2


set1 = set([1, 2, 3, 4, 5, 6])
set2 = set([4, 5, 6, 7, 8, 9])

intersection = set1 & set2 
union = set1 | set2         
symmetric_difference = set1 ^ set2  

print("Intersection:", intersection)
print("Union:", union)
print("Symmetric Difference:", symmetric_difference)

venn_diagram = venn2([set1, set2], ('Set 1', 'Set 2'))
plt.title("Venn Diagram")
plt.show()
```

Que 18:Word Count and Frequency:
        Write a program to:
        Read a file and count the number of words in it.
        Find the most frequent word.
        Visualize the word frequencies using a bar chart.

```python
import re
from collections import Counter
import matplotlib.pyplot as plt

with open("content.txt","r") as f:
    data=f.read()

our_data=re.findall(r"[a-zA-Z']+",data)
print(our_data)
counts=Counter(our_data)
x=[word for word in counts.keys()]
y=[frequency for frequency in counts.values()]
print(counts)
print("Most common term:",counts.most_common(1)[0][0])
plt.xlabel("word",size=15)
plt.ylabel("Frequency",size=15)
plt.barh(x,y,color="mediumvioletred")
plt.show()

```

Que 22:Normalization:
        Write a Python function to normalize a list of numbers to a range of 0 to 1.
        Visualize the normalized data using a line plot.

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
nums=random.randint(1,11,5)
x=np.max(nums)
y=np.min(nums)
print(nums)
normalized_nums=[]
for i in nums:
    normalized_nums.append((i-y)/(x-y))
print(normalized_nums)

plt.plot(nums,normalized_nums,color="navy",ls="-.")
plt.xlabel("Number",size=13)
plt.ylabel("Normalised number",size=13)
plt.title("Original vs Normalized Numbers",size=16)
plt.show()
```
